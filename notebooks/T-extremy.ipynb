```python
import numpy as np
import pandas as pd
```

```python
dfavgtur = pd.read_excel('../data/chmi/B2BTUR01.xlsx', sheet_name='teplota průměrná', header=3)
dfmaxtur = pd.read_excel('../data/chmi/B2BTUR01.xlsx', sheet_name='teplota maximální', header=3)
dfmintur = pd.read_excel('../data/chmi/B2BTUR01.xlsx', sheet_name='teplota minimální', header=3)
```

```python
dfavgtur.head()
```

```python
days = pd.date_range(start='01/01/1961', end='31/12/2021')
dfavgtur.index = pd.MultiIndex.from_arrays(dfavgtur[['rok', 'měsíc']].values.T)
df2 = pd.DataFrame({'date': days})
```

```python
df2.head()
```

```python
def lookupavg(dt):
    return dfavgtur.loc[dt.year, dt.month][str(dt.day) + '.']
def lookupmax(dt):
    return dfmaxtur.loc[dt.year, dt.month][str(dt.day) + '.']
def lookupmin(dt):
    return dfmintur.loc[dt.year, dt.month][str(dt.day) + '.']
```

```python
df2['tempavg'] = df2['date'].apply(lookupavg)
df2['tempmax'] = df2['date'].apply(lookupmax)
df2['tempmin'] = df2['date'].apply(lookupmin)
```
