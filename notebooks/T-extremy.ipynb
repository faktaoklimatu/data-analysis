```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('../data/chmi/B2BTUR01.xlsx', sheet_name='teplota průměrná', header=3)
```

```python
df.head()
```

```python
days = pd.date_range(start='01/01/1961', end='31/12/2021')
df.index = pd.MultiIndex.from_arrays(df[['rok', 'měsíc']].values.T)
df2 = pd.DataFrame({'date': days})
def lookup(dt):
    return df.loc[dt.year, dt.month][str(dt.day) + '.']
```

```python
df2['temp'] = df2['date'].apply(lookup)
```

```python
df2.head()
```

```python
import seaborn as sns

sns.lineplot(data=df2.iloc[:365], x='date', y='temp')
```
